```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

from augment.geometric import crop
```

```python
img = cv2.imread('images/0.jpeg')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

img, box = crop(img, (700, 400), (800, 450), [581, 274, 699, 321])
print(box)
cv2.rectangle(img, (box[0], box[1]), (box[2], box[3]), (0, 255, 0), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img = cv2.imread('images/0.jpeg')
h, w = img.shape[:2]
M = cv2.getRotationMatrix2D((w / 2, h / 2), 410.5, 1.0)
img_new = cv2.warpAffine(img, M, (w, h))
plt.imshow(img_new)
```
